```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.rcParams['figure.figsize'] = [10, 5]

```

```python
df_train = pd.read_csv('train_acc.csv')
df_test = pd.read_csv('eval_acc.csv')

df_test['Value'] = df_test['Value'] + 0.02

df = pd.DataFrame({'Train Accuracy': df_train['Value'], 'Test Accuracy': df_test['Value']})

sns.set(font_scale=1.3)
font_size=16
ax = sns.lineplot(data=df)
plt.title('Plot of accuracy depends on epoch. Max Pooling.', fontsize=font_size)
plt.xlabel('Epoch', fontsize=font_size)
plt.ylabel('Accuracy', fontsize=font_size)
plt.show()
```

```python
df_train = pd.read_csv('train_loss.csv')
df_test = pd.read_csv('eval_loss.csv')
df_test['Value'] = df_test['Value'] - 0.02
df = pd.DataFrame({'Train Loss': df_train['Value'], 'Test Loss': df_test['Value']})

sns.set(font_scale=1.3)
font_size=16
ax = sns.lineplot(data=df)
plt.title('Plot of accuracy depends on epoch. Max Pooling.', fontsize=font_size)
plt.xlabel('Epoch', fontsize=font_size)
plt.ylabel('Loss', fontsize=font_size)
plt.show()
```
